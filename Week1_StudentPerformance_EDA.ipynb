```python
# These are the tools we need for this project
import pandas as pd              # To work with data
import matplotlib.pyplot as plt  # To make graphs
import seaborn as sns            # To make prettier graphs

# Show graphs inside notebook
%matplotlib inline





```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()



```

```python

df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python

df.groupby('gender')[['math score', 'reading score', 'writing score']].mean()
```

```python

sns.barplot(x='gender', y='math score', data=df)
plt.title('Average Math Score by Gender')
plt.show()
```

```python

sns.boxplot(x='test preparation course', y='writing score', data=df)
plt.title('Writing Score by Test Preparation Course')
plt.show()
```

```python

plt.hist(df['reading score'], bins=10, color='skyblue')
plt.title('Distribution of Reading Scores')
plt.xlabel('Reading Score')
plt.ylabel('Number of Students')
plt.show()
```

Female students score higher in reading and writing.

Students who took test preparation scored better overall.

Students with standard lunch tend to score higher than those with free/reduced lunch.

Average Math scores are slightly higher for males.

Reading scores mostly range between 5080.
